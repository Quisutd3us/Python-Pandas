```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/d3us/Desktop/cursoDataScienceLinkedIn/bd/2008.csv", nrows=1e6)
df.head()
```

```python
df2= df.copy()
```

```python
df2= df2.sample(frac=1)
```

```python
#eliminando duplicados
df2clean= df2.drop_duplicates()
len(df2clean) == len(df2)
```

```python
texto='{0} -> {1}'.format(len(df2),len(df2clean))
print(texto)
```

```python
#borra los valores duplicados de la columna DayofMonth , en este caso me deja una bd con 31 dias o 31 registros
df2clean.drop_duplicates(subset="DayofMonth")
```

```python
#borrar NaN -> usando el parametro thresh le decimos cuantos columnas llenas de las que hay
df2clean.dropna(thresh=len(df2clean.columns)-2)
```

```python
df3= df.copy()
df3.head()
```
